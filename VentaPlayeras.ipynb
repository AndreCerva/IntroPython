#### Vendedor de playera de 4 equipos, chivas, america, pumas, atlas, c/u cuesta $100

##### Situación 1:
###### Chivas - 10%, América -20% otro club no hay descuento 

```python
total=100
descuentochivas=0.90
descuentoamerica=0.80
def main():
    while True:
        playera=input("Introduzca la playera que desea: ")
        if playera.lower()=="chivas" or int(playera)==1:
            print(f"""La playera del chivas tiene un descuento del 10%!
                       El precio total a pagar es: {total*descuentochivas}""")
        elif playera.lower()=="america" or int(playera)==2:
            print(f"""La playera del america tiene un 20% de descuento!
                      El precio total a pagar es: {total*descuentoamerica}""")
        elif playera.lower()=="pumas" or int(playera)==3 or playera.lower()=="atlas" or int(playera)==4:
            print(f"El total por la playera {playera} es: {total}")
        elif playera.lower()=="salir" or int(playera)==5:
            print("Adiós!")
            break
        else:
            print(f"No tenemos la playera {playera} en existencia :c")

if __name__ =='__main__':
    print("""Las siguientes playeras son las que tenemos disponibles: 
    1) Chivas
    2) America
    3) Pumas
    4) Atlas
    5) Salir""")
    main()
```
